# Use for classifcation 
Given a set of training in BINARY - SVM will assign new test poitns into one categroy or the other.
- 'Shows a division of a clear gap as much as possible'
- maximizze the hyperplace distances margins from all the points


```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat
y = cancer['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
print(classification_report(y_test, predictions))
```

# GRID SEARCH

```python
from sklearn.grid_search import GridSearchCV
```

I believe the best way to udnerstand this is:
- A SVM has a linear HYERPLANE which is drawn to seperate the data. However, sometimes the BEST line for predictions does not invovle a PERFECT seperation (C=0). Thus, we change values of C to allow an increase in tolerance to how many VIOLATIONS to the MARGIN / HYPERPLANE we can make
- Where as GAMMA is how non linear something gets, a larger gamma will haev ahigher bias.

```python
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
confusion_matrix(y_test, grid_predictions)
```

```python
print(classification_report(y_test, grid_predictions))
```
